```python
#Here we use fish dataset where each different fish has different length,height,width..as we are using this dataset for linear regression with multiple non dependent values, we are going to use weight as output to be predicted using other features..we gonna drop species column as we dont have any use here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Fish.csv")
df.head(5)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
reg=linear_model.LinearRegression()
x=df.drop(columns=["Species","Weight"])
y=df["Weight"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
reg.fit(x_train,y_train)
prediction=reg.predict(x_test)
print("the predicted weight is:",prediction)
```

```python
plt.xlabel("actual weight of the fish")
plt.ylabel("predicted weight of the fish")
plt.title("linear regression algorithm")
plt.scatter(y_test,prediction)
plt.show()
```

```python
#r squared score cheking for regression
accuracy=r2_score(y_test,prediction)
accuracy1=accuracy*100
print("the R Squared score of the model is:",accuracy1)
```
