```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_csv("Fish dataset.csv")
print(df.head(10))
```

```python
#plotting
plt.scatter(df["Species"],df["Weight"])
plt.xlabel("species")
plt.ylabel("features")
plt.show()
```

```python
x=df.drop("Species",axis=1)
y=df["Species"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
predicted=model.predict(x_test)
print(predicted)
```

```python
#testing
# Create a DataFrame with a single row, specifying an index
nw = pd.DataFrame({"Weight":[234],"Length1":[234],"Length2":[234],"Length3":[234],"Height":[234], "Width":[234]}, index=[0])
print(model.predict(nw))
```
