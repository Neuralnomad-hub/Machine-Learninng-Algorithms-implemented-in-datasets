Problem Statement

Predict the forest cover type (tree species/ecosystem) of a 30Ã—30 meter area of forest land based on environmental and geographic features such as elevation, slope, soil type, and distance to water/roads/fires.

Target Classes (Cover_Type)

1 â†’ Spruce/Fir

2 â†’ Lodgepole Pine

3 â†’ Ponderosa Pine

4 â†’ Cottonwood/Willow

5 â†’ Aspen

6 â†’ Douglas-fir

7 â†’ Krummholz

ðŸ‘‰ So the task = Multiclass classification (7 classes) predicting the type of forest cover.

```python
from sklearn.datasets import fetch_covtype
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=fetch_covtype(as_frame=True)
df = df.frame
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python
x=df.drop('Cover_Type',axis=1)
y=df['Cover_Type']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
print(model.predict(x_test))
print(classification_report(y_test,model.predict(x_test)))
```
